```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Step1 Load the data

```python
df = pd.read_csv('auto.txt')
df.head()
```

## Stpe 2: cleaning data

```python
import numpy as np

df = df.replace('?',np.nan)
df
```

```python
df.dtypes
```

```python
df['make'].unique()

make_dict = {
    'make':{
        'alfa-romero' :1,
        'audi' :2,
        'bmw' :3,
        'chevrolet' :4,
        'dodge' : 5,
        'honda' : 6,
        'isuzu' : 7,
        'jaguar' : 8, 
        'mazda' : 9, 
        'mercedes-benz' :10,
        'mercury' : 11,
       'mitsubishi' : 12, 
        'nissan' :13, 
        'peugot' :14, 
        'plymouth' : 15, 
        'porsche' : 16,
        'renault' :17,
       'saab' :18,
        'subaru' :19, 
        'toyota' :20, 
        'volkswagen' :21, 
        'volvo' :22
    }
}
df.replace(make_dict,inplace=True)
df['make'] = pd.to_numeric(df['make'])
```

```python
df['fuel-type'].unique()
fuel_type = {
    'gas':1,
    'diesel':2
}
df['fuel-type'].replace(fuel_type,inplace=True)
df['fuel-type'] = pd.to_numeric(df['fuel-type'])
```

```python
df['aspiration'].unique()
asp_dict ={
    'aspiration':{
        'std' :1,
        'turbo' :2
    }
}
df.replace(asp_dict,inplace=True)
df['aspiration'] = pd.to_numeric(df['aspiration'])
```

```python
df['num-of-doors'].unique()
num_doors = {
    'four':4,
    'two':2
}
df['num-of-doors'].replace(num_doors,inplace=True)
df['num-of-doors'] = pd.to_numeric(df['num-of-doors'])
```

```python
df['body-style'].unique()

df.replace(df['body-style'].unique(),
           range(1,len(df['body-style'].unique())+1),
           inplace=True)
```

```python
df['drive-wheels'].unique()
df.replace(df['drive-wheels'].unique(),
          range(1,len(df['drive-wheels'].unique())+1),
           inplace=True)
```

```python
df['engine-location'].unique()
df.replace(df['engine-location'].unique(),
          range(1,len(df['engine-location'].unique())+1),
          inplace=True)
```

```python
dict = {
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'twelve':12
}
df['num-of-cylinders'].replace(dict,inplace=True)
df['num-of-cylinders'] = pd.to_numeric(df['num-of-cylinders'])
```

```python
engine_type = {
    'dohc':1,
    'dohcv':2,
    'l':3,
    'ohc':4,
    'ohcf':5,
    'ohcv':6,
    'rotor':7
}
df['engine-type'].replace(engine_type,inplace=True)
df['engine-type'] = pd.to_numeric(df['engine-type'])
```

```python
fule_sys = {
    '1bbl':1,
    '2bbl':2,
    '4bbl':4,
    'idi':5,
    'mfi':6,
    'mpfi':7,
    'spdi':8,
    'spfi':9
}
df['fuel-system'].replace(fule_sys,inplace=True)
df['fuel-system'] = pd.to_numeric(df['fuel-system'])
```

```python
df['price'] = pd.to_numeric(df['price'],errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'],errors='coerce')
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'],errors='coerce')
df['bore'] = pd.to_numeric(df['bore'],errors='coerce')
df['stroke'] = pd.to_numeric(df['stroke'],errors = 'coerce')
```

```python
df.dtypes
```

```python
df.drop('symboling',axis=1,inplace=True)
df.drop('normalized-losses',axis=1,inplace=True)
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
fillna_dict = {
    'num-of-doors' : df['num-of-doors'].mean(),
    'bore' :df['bore'].mean(),
    'stroke' : df['stroke'].mean(),
    'horsepower' :df['horsepower'].mean(),
    'peak-rpm' :df['peak-rpm'].mean(),
    'price' :df['price'].mean()
}
df.fillna(fillna_dict,inplace=True)
```

```python
df.drop(['horsepower','curb-weight','peak-rpm','length','city-mpg'],inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split

inputs_coulmns = df.drop('price',axis=1)
output_coulmns = df['price']

x_train,x_test,y_train,y_test = train_test_split(inputs_coulmns,output_coulmns,random_state=0,test_size=0.2)
```

## Step 3: Training and Testing

```python
from sklearn.linear_model import LinearRegression
lmodel = LinearRegression()
lmodel.fit(x_train,y_train.ravel())
```

```python
# lmodel.score(x_test,y_test)
```

```python
pd.Series(lmodel.coef_,x_train.columns).sort_values()
```

```python
predict_value = lmodel.predict(x_test)
plt.figure(figsize=(10,5))
plt.plot(y_test.values,label='Actual Value')
plt.plot(predict_value,label= 'Predicted value')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
liner_mse = mean_squared_error(predict_value,y_test) 
liner_mse
```
